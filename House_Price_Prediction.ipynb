```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset=sklearn.datasets.house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
house_price_df=pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
```

```python
house_price_df.head()
```

```python
house_price_df['price']=house_price_dataset.target
```

```python
house_price_df.head()
```

```python
house_price_df.shape
```

```python
#check for missing value
```

```python
house_price_df.isnull().sum()
```

```python
house_price_df.describe()
```

Understanding Correlation

```python
correlation=house_price_df.corr()
```

```python


# Heatmap
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Reds')
```

```python
#Spliting data and targets
X=house_price_df.drop(['price'],axis=1)
Y=house_price_df['price']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model XGBoost Regressor

```python
model=XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

Evaluation

```python
training_prediction=model.predict(X_train)
```

```python
print(training_prediction)
```

```python
#R squared Error
score_1=metrics.r2_score(Y_train,training_prediction)

#Mean absolute Error
score_2=metrics.mean_absolute_error(Y_train,training_prediction)
print('R2 Error : ',score_1)
print('Mean Ab Error : ',score_2)
```

```python
plt.scatter(Y_train, training_prediction, color='red')
plt.scatter(Y_train, Y_train, color='blue')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.show()
```

```python
testing_prediction=model.predict(X_test)
print(testing_prediction)
```

```python
#R squared Error
score_1=metrics.r2_score(Y_test,testing_prediction)

#Mean absolute Error
score_2=metrics.mean_absolute_error(Y_test,testing_prediction)
print('R2 Error : ',score_1)
print('Mean Ab Error : ',score_2)
```

```python
plt.scatter(Y_test, testing_prediction, color='red')
plt.scatter(Y_test, Y_test, color='blue')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.show()
```
